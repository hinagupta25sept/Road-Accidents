```
df = {}
df["Driver's Fault"] = data.loc[36,'Fault of Driver-Total No. of Road Accidents - 2014'] 
df["Cyclist's Fault"] = data.loc[36,'Fault of Cyclist-Total No. of Road Accidents - 2014'] 
df["Vehicle Condition"] = data.loc[36,'Defect in Condition of Motor Vehicle-Total No. of Road Accidents - 2014'] 
df["Road Condition"] = data.loc[36,'Defect in Road Condition-Total No. of Road Accidents - 2014'] 
df["Weather Condition"] = data.loc[36,'Weather Condition-Total No. of Road Accidents - 2014'] 
df["Passenger's Fault"] = data.loc[36,'Fault of Passenger-Total No. of Road Accidents - 2014'] 
df["Poor Light"] = data.loc[36,'Poor light-Total No. of Road Accidents - 2014'] 
df["Stray Animals"] = data.loc[36,'Stray animals-Total No. of Road Accidents - 2014'] 
df["Others"] = data.loc[36,'Other causes/ Causes not known-Total No. of Road Accidents - 2014'] 
```

```
val = list(df.values())
total = sum(val)
for i in range(0,9):
    val[i] = format(val[i]*100/total,'.2f')

plt.figure(figsize=(10,8))
plt.pie(list(df.values()))
plt.axis('equal')
plt.xlabel('Accidents in 2014')

centre_circle = plt.Circle((0,0),0.70,fc='grey')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
label = [list(pair) for pair in zip(list(df.keys()),val)]
plt.legend(label,loc="best")
plt.show()
```

**It can be infered from the above graph that main cause of Accidents in India is Driver's Fault.**
